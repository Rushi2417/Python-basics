```python
with open("score.csv", "r", encoding="utf-8-sig") as f:
    d = {}
    for line in f:
        line = line.strip()
        name, _, sco = line.split(",")
        sco = int(sco)
        if name in d:
            d[name].append(sco)
        else:
            d[name] = [sco]
    print(d)

    for players, scores in d.items():
        print(players, min(scores), max(scores), sum(scores)/3)
```
